## í€˜ìŠ¤íŠ¸ 1ë²ˆ: ì€í–‰ ê³„ì¢Œ ë§Œë“¤ê¸°

````python
import random
import time

class Account:
    account_num = 0
    
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.bank = "SCì€í–‰"

        x = random.randint(100,999)
        y = random.randint(10,99)
        z = random.randint(100000,999999)
        self.account = str(x) + '-' + str(y) + '-' + str(z)

        self.deposit_num = 0
        Account.account_num += 1

        # ì…ê¸ˆ ê¸°ë¡ ì €ì¥ (ì‹ ê·œ ê³„ì¢Œê°œì„¤ì‹œ ì…ê¸ˆ)
        self.deposit_h = []
        self.withdraw_h = []
        t = time.localtime()
        formatted = time.strftime("%Y-%m-%d %H:%M:%S", t)
        self.deposit_h.append(f"ê³„ì¢Œê°œì„¤  ì‹œê°„: {formatted},  ì…ê¸ˆê¸ˆì•¡: {balance}")

    # ê³„ì¢Œì˜ ê°œìˆ˜ ì¶œë ¥
    def get_account_num(self):
        return Account.account_num

    # ì…ê¸ˆ
    def deposit(self):
        while(True):
            money = int(input("ì…ê¸ˆ ê¸ˆì•¡ì„ ì…ë ¥í•˜ì„¸ìš”. ì…ê¸ˆì€ 1ì›” ì´ìƒë§Œ ê°€ëŠ¥í•©ë‹ˆë‹¤.\n >>> "))
            if money < 1:
                print("ì…ê¸ˆì€ ìµœì†Œ 1ì› ì´ìƒë¶€í„° ê°€ëŠ¥í•©ë‹ˆë‹¤.")
                n = int(input("ë‹¤ì‹œ ì…ê¸ˆì€ 1, ì¢…ë£ŒëŠ” 0ì„ ì…ë ¥í•˜ì„¸ìš”.\n >>> "))
                if n == 0:
                    return
                        
            else:
                self.balance += money
                self.deposit_num += 1

                # ì…ê¸ˆ ê¸°ë¡ ì €ì¥
                t = time.localtime()
                formatted = time.strftime("%Y-%m-%d %H:%M:%S", t)
                self.deposit_h.append(f"ì¶”ê°€ì…ê¸ˆ  ì‹œê°„: {formatted},  ì…ê¸ˆê¸ˆì•¡: {money}")

                # ì…ê¸ˆ íšŒìˆ˜ê°€ 5íšŒê°€ ë  ë•Œ ì”ê³ ë¥¼ ê¸°ì¤€ìœ¼ë¡œ 1%ì˜ ì´ì ì¶”ê°€
                if self.deposit_num == 5:
                    self.balance *= 1.01
                
                return

    # ì¶œê¸ˆ
    def withdraw(self):
        while(True):
            money = int(input("ì¶œê¸ˆ ê¸ˆì•¡ì„ ì…ë ¥í•˜ì„¸ìš”.\n >>> "))
            if money > self.balance:
                print(f"ì¶œê¸ˆì€ ê³„ì¢Œì˜ ì”ê³  ì´ìƒìœ¼ë¡œ ì¶œê¸ˆí•  ìˆ˜ëŠ” ì—†ìŠµë‹ˆë‹¤. ì”ê³ : {self.balance:,}")
                n = int(input("ë‹¤ì‹œ ì¶œê¸ˆì€ 1, ì¢…ë£ŒëŠ” 0ì„ ì…ë ¥í•˜ì„¸ìš”.\n >>> "))
                if n == 0:
                    return
                        
            else:
                self.balance -= money

                # ì¶œê¸ˆ ê¸°ë¡ ì €ì¥
                t = time.localtime()
                formatted = time.strftime("%Y-%m-%d %H:%M:%S", t)
                self.withdraw_h.append(f"ì¶œê¸ˆ  ì‹œê°„: {formatted},  ì¶œê¸ˆê¸ˆì•¡: {money}")
                
                return

    # ê³„ì¢Œ ì •ë³´ ì¶œë ¥
    def display_info(self):
        print(f"ì€í–‰ì´ë¦„: {self.bank}, ì˜ˆê¸ˆì£¼: {self.name}, ê³„ì¢Œë²ˆí˜¸: {self.account}, ì”ê³ : {self.balance:,}")

    # ì…ê¸ˆ ë‚´ì—­ ì¶œë ¥
    def deposit_history(self):
        for history in self.deposit_h:
            print(history)

    # ì¶œê¸ˆ ë‚´ì—­ ì¶œë ¥
    def withdraw_history(self):
        for history in self.withdraw_h:
            print(history)
````

## í€˜ìŠ¤íŠ¸ 2ë²ˆ: ê°„ë‹¨í•œ ìë™ì‚¬ëƒ¥ RPG ë§Œë“¤ê¸°

````python
import random

class Character:
    def __init__(self, name):
        self.name = name
        self.level = 0
        self.strength = 0
        self.attack = 0
        self.defense = 0

    def is_alive(self):
        return self.strength >= 0

    # ê³µê²© ë°›ì•˜ì„ ë•Œ
    def take_damage(self, r_attack):
        damage = r_attack - self.defense
        if damage > 0:
            self.strength -= damage
            return damage
        else:
            return 0

    # íƒ€ê²Ÿì—ê²Œ ë°ë¯¸ì§€ ì…íˆê¸°
    def attack_target(self):
        if self.attack > 0:
            t_attack = random.randint(1, self.attack)
            return t_attack
        else:
            return 0

    # ì •ë³´ì¶œë ¥
    def display_info(self):
        print(f"ì´ë¦„: {self.name}, ë ˆë²¨: {self.level}, ì²´ë ¥: {self.strength}, ê³µê²©ë ¥: {self.attack}, ë°©ì–´ë ¥: {self.defense}")

class Player(Character):
    def __init__(self, name):
        super().__init__(name)
        self.level = 1
        self.strength = 100
        self.attack = 25
        self.defense = 5
        self.experience = 0

    def gain_experience(self, num):
        self.experience += num

    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            print("level up")

    # ì •ë³´ì¶œë ¥
    def display_info(self):
        print(f"ì´ë¦„: {self.name}, ë ˆë²¨: {self.level}, ì²´ë ¥: {self.strength}, ê³µê²©ë ¥: {self.attack}, ë°©ì–´ë ¥: {self.defense}, ê²½í—˜ì¹˜: {self.experience}")
         
class Monster(Character):
    def __init__(self, name, level):
        super().__init__(name)
        self.level = level
        self.strength = random.randint(10,30) * self.level
        self.attack = random.randint(5,20) * self.level
        self.defense = random.randint(1,5) * self.level

def battle(player, monster):
    player.display_info()
    monster.display_info()
    print()
    
    while(True):
        player_power = player.attack_target()
        m_damage = monster.take_damage(player_power)

        monster_power = monster.attack_target()
        p_damage = player.take_damage(monster_power)

        print("ì´ë¦„: ", player.name, " : ë°›ì€ ê³µê²©: ", monster_power, "damage: ", p_damage, " ë‚¨ì€ Playerì²´ë ¥: ", player.strength)
        print("ì´ë¦„: ", monster.name, " : ë°›ì€ ê³µê²©: ", player_power, "damage: ", m_damage, " ë‚¨ì€ Monsterì²´ë ¥: ", monster.strength)

        if monster.strength <= 0 or player.strength <= 0:
            break

    if player.strength > 1:
        print("ì „íˆ¬ìŠ¹ë¦¬!")
        player.experience += monster.level * 20
        player.level_up()
        return 0
    else:
        print("ì „íˆ¬ íŒ¨ë°°..")
        return 1

def main():
    monster_dict = {'ìŠ¬ë¼ì„': 1, 'ê³ ë¸”ë¦°': 2, 'ì˜¤í¬': 3}
    
    name = input("Player ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” :")
    player = Player(name)
    win = 0
    
    for name, level in monster_dict.items():
        monster = Monster(name, level)
        if battle(player, monster):
            win = 1
            break

    if win:
        print("ê²Œì„ì˜¤ë²„")
    else:
        print("ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì´ê²¼ë‹¤!")
````

````python
main()
````

(ì…ê¸ˆ ë‚´ì—­ê³¼ ì¶œê¸ˆ ë‚´ì—­ì„ ì¶œë ¥í•˜ëŠ” deposit_historyì™€ withdraw_history ë©”ì„œë“œë¥¼ ì¶”ê°€í•˜ì„¸ìš”.)

````python
import random


class Account:
    # Q2. í´ë˜ìŠ¤ ë³€ìˆ˜ (ê³„ì¢Œ ê°œìˆ˜)
    account_count = 0

    def __init__(self, owner, balance):
        # Q1. ìƒì„±ì
        self.bank = "SCì€í–‰"
        self.owner = owner
        self.balance = balance

        # ê³„ì¢Œë²ˆí˜¸ ëœë¤ ìƒì„± (3-2-6)
        self.account_number = (
            f"{random.randint(100, 999)}-"
            f"{random.randint(10, 99)}-"
            f"{random.randint(100000, 999999)}"
        )

        # Q7. ì…ê¸ˆ íšŸìˆ˜
        self.deposit_count = 0

        # Q10. ì…ì¶œê¸ˆ ë‚´ì—­
        self.deposit_list = []
        self.withdraw_list = []

        # ê³„ì¢Œ ê°œìˆ˜ ì¦ê°€
        Account.account_count += 1

    # Q3. í´ë˜ìŠ¤ ë³€ìˆ˜ ì¶œë ¥
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. ì…ê¸ˆ ë©”ì„œë“œ
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_list.append(amount)

            # Q7. ì…ê¸ˆ 5íšŒë§ˆë‹¤ ì´ì ì§€ê¸‰ (1%)
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
        else:
            print("ì…ê¸ˆ ê¸ˆì•¡ì€ 1ì› ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤.")

    # Q5. ì¶œê¸ˆ ë©”ì„œë“œ
    def withdraw(self, amount):
        if amount > self.balance:
            print("ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤.")
        else:
            self.balance -= amount
            self.withdraw_list.append(amount)

    # Q6. ì •ë³´ ì¶œë ¥ ë©”ì„œë“œ
    def display_info(self):
        print(f"ì€í–‰ì´ë¦„: {self.bank}")
        print(f"ì˜ˆê¸ˆì£¼: {self.owner}")
        print(f"ê³„ì¢Œë²ˆí˜¸: {self.account_number}")
        print(f"ì”ê³ : {self.balance:,}ì›")

    # Q10. ì…ê¸ˆ ë‚´ì—­ ì¶œë ¥
    def deposit_history(self):
        print("ì…ê¸ˆ ë‚´ì—­:")
        for money in self.deposit_list:
            print(f"  +{money:,}ì›")

    # Q10. ì¶œê¸ˆ ë‚´ì—­ ì¶œë ¥
    def withdraw_history(self):
        print("ì¶œê¸ˆ ë‚´ì—­:")
        for money in self.withdraw_list:
            print(f"  -{money:,}ì›")
````

````python
import random


# ===============================
# Q1. Character í´ë˜ìŠ¤
# ===============================
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    # ìƒì¡´ ì—¬ë¶€ í™•ì¸
    def is_alive(self):
        return self.hp > 0

    # ë°ë¯¸ì§€ ë°›ê¸°
    def take_damage(self, damage):
        real_damage = damage - self.defense
        if real_damage > 0:
            self.hp -= real_damage
        if self.hp < 0:
            self.hp = 0

    # ê³µê²©
    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        target.take_damage(damage)
        return damage


# ===============================
# Q2. Player í´ë˜ìŠ¤
# ===============================
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.exp = 0

    # ê²½í—˜ì¹˜ íšë“
    def gain_experience(self, amount):
        self.exp += amount

    # ë ˆë²¨ì—…
    def level_up(self):
        while self.exp >= 50:
            self.exp -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            print(f"ğŸ‰ ë ˆë²¨ ì—…! â†’ Lv.{self.level}")


# ===============================
# Q2. Monster í´ë˜ìŠ¤
# ===============================
class Monster(Character):
    def __init__(self, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        super().__init__(
            name=f"ëª¬ìŠ¤í„°(Lv.{level})",
            level=level,
            hp=hp,
            attack=attack,
            defense=defense
        )


# ===============================
# Q3. battle í•¨ìˆ˜
# ===============================
def battle(player, monster):
    print(f"\nâš”ï¸ ì „íˆ¬ ì‹œì‘! {player.name} vs {monster.name}")

    while player.is_alive() and monster.is_alive():
        # í”Œë ˆì´ì–´ ê³µê²©
        dmg = player.attack_target(monster)
        print(f"{player.name} â†’ {monster.name} ê³µê²© ({dmg} ë°ë¯¸ì§€)")
        print(f"{monster.name} HP: {monster.hp}")

        if not monster.is_alive():
            break

        # ëª¬ìŠ¤í„° ê³µê²©
        dmg = monster.attack_target(player)
        print(f"{monster.name} â†’ {player.name} ê³µê²© ({dmg} ë°ë¯¸ì§€)")
        print(f"{player.name} HP: {player.hp}")

    # ì „íˆ¬ ê²°ê³¼
    if player.is_alive():
        print("âœ… ì „íˆ¬ ìŠ¹ë¦¬!")
        exp = monster.level * 20
        player.gain_experience(exp)
        print(f"ê²½í—˜ì¹˜ {exp} íšë“!")
        player.level_up()
    else:
        print("âŒ ì „íˆ¬ íŒ¨ë°°..")


# ===============================
# ê²Œì„ ì‹¤í–‰
# ===============================
if __name__ == "__main__":
    player_name = input("í”Œë ˆì´ì–´ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    player = Player(player_name)

    print(f"\nğŸ® {player.name}ì˜ ìë™ì‚¬ëƒ¥ ì‹œì‘!\n")

    for level in range(1, 4):
        monster = Monster(level)
        battle(player, monster)

        if not player.is_alive():
            print("\nğŸ’€ ê²Œì„ ì˜¤ë²„!")
            break
    else:
        print("\nğŸ† ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì²˜ì¹˜í–ˆìŠµë‹ˆë‹¤!")
````
