## 프로젝트 1. 사각형 넓이 구하는 클래스 생성

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    # 직사각형 넓이
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    # 평행사변형 넓이
    def par(self):
        lower, height = map(int, input("밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>").split(','))
        area = lower * height
        result = "평행사변형의 넓이는 : " + str(area)
        return result

    # 사다리꼴 넓이
    def trape(self):
        lower, upper, height = map(int, input("밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>").split(','))
        area = (lower + upper) * height * 1/2
        result = "사다리꼴의 넓이는 : " + str(area)
        return result
```

```python
a = Square()
```

```python
a.rect()
```

```python
b = Square()
```

```python
b.par()
```

```python
c = Square()
```

```python
c.trape()
```

## 프로젝트 2. Kiosk 클랙스 생성

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.ordered_menu = [] # 주문한 메뉴 리스트
        self.ordered_price = [] # 주문한 가격 리스트

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(f'No: {i+1}. {self.menu[i]}  :  {self.price[i]} 원')

    # 주문 메서드
    def menu_select(self):

        # 추가 주문이 없을 때까지 반복 
        n = 1
        while(n != 0) :
            # 음료 선택
            menu_select = 0
            while menu_select < 1 or len(self.menu) < menu_select:
                print()
                menu_select = int(input('음료 번호를 입력하세요 : '))

                # 메뉴판에 없는 번호일때
                if menu_select < 1 or len(self.menu) < menu_select:
                    print("없는 메뉴입니다. 다시 주문해 주세요")
            
            # 음료 온도 물어보기
            temp_select = 0
            while temp_select != 1 and temp_select != 2:
                temp_select = int(input("HOT 음료는 1,  ICE 음료는 2를 입력하세요 : "))

                if temp_select == 1:
                    self.temp = "HOT"
                elif temp_select == 2:
                    self.temp = "ICE"
                else:
                    print("1과 2중 하나를 입력하세요 (1. HOT, 2. ICE)\n")

            # 주문 결과 출력
            print('\n주문 음료', self.temp, self.menu[menu_select-1], ' : ', self.price[menu_select-1], '원')

            # 주문 결과 저장
            self.ordered_price.append(self.price[menu_select-1])
            self.ordered_menu.append(self.temp + ' ' + self.menu[menu_select-1])

            # 현재까지 주문 합계 출력
            print('\n총 주문 금액 : ', sum(self.ordered_price), '원')

            # 추가 주문 문의
            n = int(input("\n추가 주문을 원하시면 1, 종료는 0을 입력하세요. "))

            # 종료전 최종 주문 결과 출력
            if n == 0:
                print("\n\n최종 주문 결과 :")
                for i in range(len(self.ordered_menu)):
                    print(self.ordered_menu[i], " : ", self.ordered_price[i])
                print("\n 합계 : ", sum(self.ordered_price))

    # 지불
    def pay(self):
        # 지불할 금액 출력
        print(f"\n총 지불할 금액은 {sum(self.ordered_price)}원 입니다.")

        # 지불 수단 문의
        select = 0
        while select != 1 and select != 2:  # 1이나 2를 입력할 때까지 물어보기
            select = int(input("\n지불 수단을 선택해주세요.\n 현금은 1, 카드는 2를 입력하세요 : "))
            if select == 1:
                print("\n\n현금 → 직원을 호출하겠습니다.\n")
            elif select == 2:
                print("\n\n카드 → IC칩 방향에 맞게 카드를 꽂아주세요.\n")
            else:    
                print("\n\n1과 2 중 하나를 입력하세요.\n")
       

    # 주문서 출력 
    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        print("주문 내역 : ")
        for i in range(len(self.ordered_menu)):
            print(self.ordered_menu[i], ' : ', self.ordered_price[i])

        print('\n합계 : ', sum(self.ordered_price))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
